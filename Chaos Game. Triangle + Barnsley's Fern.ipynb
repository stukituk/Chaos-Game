```python
def middle(dot1, dot2):
    return [0.5*(dot1[0]+dot2[0]), 0.5*(dot1[1]+dot2[1])]
```

```python
from scipy import sqrt, zeros
import matplotlib.pyplot as plt
from random import randint
A = [0,0]
B = [1,0]
C = [0.5, 1]
N = 10000
x = [0 for i in range(0,N)]
y = [0 for i in range(0,N)]
for i in range(1, N):
    rand = randint(0,2)
    if rand == 0:
        x[i], y[i] = middle([x[i-1], y[i-1]], A)
    elif rand == 1:
        x[i], y[i] = middle([x[i-1], y[i-1]], B)
    else:
        x[i], y[i] = middle([x[i-1], y[i-1]], C)
plt.scatter(x, y, s = 1)
plt.show()
```
